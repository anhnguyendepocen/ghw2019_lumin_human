```python
from pathlib import Path
import fiona
import geopandas as gpd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_pth = Path("../data")
```

```python
# Load Natural Earth Data for Africa
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
countries = world[world["continent"] == "Africa"]
countries.plot(figsize = (8,8))

# Get coordinate system of data
countries.crs
```

```python
# Save as shapefile and geojson
# To identify all supported file formats: fiona.supported_drivers
countries[countries["name"] == "Tanzania"].to_file("tanzania.shp", driver="Shapefile")
countries[countries["name"] == "Nigeria"].to_file("nigeria.shp", driver="Shapefile")

# Entire African continent; GeoJSONs don't like data frames with heterogeneous geometries -- In this instance, there are both multipolygons and polygons
#countries.to_file("african_countries.shp", driver="Shapefile")
```
